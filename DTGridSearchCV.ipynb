```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid={'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
            'splitter':['best', 'random'],'max_features':['int', 'float','sqrt', 'log2']}
grid=GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
result=grid.cv_results_
```

```python
result
```

```python

table=pd.DataFrame.from_dict(result)
```

```python
table
```

```python
table.columns
table=table.sort_values(by='rank_test_score')
table
```

```python
dttable=table[['param_criterion','param_splitter','std_test_score','rank_test_score']]
dttable.columns=[['Criterion','Splitter','STD Score','Rank']]
dttable

```

```python
Age = float(input("Please enter your age: "))
BMI = float(input("Please enter your BMI: "))
Children = float(input("Please enter number of children: "))

Gender = input("Please enter your gender: ").lower()
if Gender == "male":
    gender = 1
else:
    gender = 0

Smoker = input("Do you smoke? (yes/no): ").lower()
if Smoker == "yes":
    smoker = 1
else:
    smoker = 0

charger_prediction = grid.predict([[Age, BMI, Children, gender, smoker]])

print("Estimated Charge is:", charger_prediction[0])
```
