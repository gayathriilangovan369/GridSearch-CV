```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'C':[10,100,1000,2000,3000],'gamma':['scale','auto']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
result=grid.cv_results_
```

```python
result
```

```python
table=pd.DataFrame.from_dict(result)
table=table.sort_values(by='rank_test_score')
svr_gridsearch_results = table[[
    'param_C',
    'param_gamma',
    'param_kernel',
    'mean_test_score',
    'rank_test_score'
]]
svr_gridsearch_results.columns = ['C', 'gamma', 'kernel', 'score', 'rank']
svr_gridsearch_results
```
