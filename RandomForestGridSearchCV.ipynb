```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[[ 'charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid={'n_estimators':[10,20,30,40,50,100],'criterion':['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],'bootstrap':[True]}
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
result=grid.cv_results_
result
```

```python
table=pd.DataFrame.from_dict(result)
table
```

```python
table.columns
```

```python
table=table.sort_values(by='rank_test_score')
table
```

```python
table.columns
```

```python
rdtable=table[['param_criterion', 'param_n_estimators', 'param_bootstrap','std_test_score', 'rank_test_score','mean_test_score']]
rdtable.columns=[['Criterion','N_Estimators','Bootstrap','SDScore','Mean_Test_Score','Rank']]
rdtable
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2score=r2_score(y_test,y_pred)
r2score
```
